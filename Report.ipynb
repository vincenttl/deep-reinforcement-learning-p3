Learning Algorithm
The agent is using basic maddpg algorithm to solve the tennis task. It uses a shared replay buffer to store past experience and uses batches of experiences to train the model.

List of hyperparameters:

BUFFER_SIZE = int(1e6) # replay buffer size  
BATCH_SIZE = 512 # minibatch size  
GAMMA = 0.99 # discount factor  
TAU = 2e-1 # for soft update of target parameters  
LR_ACTOR = 1e-4 # learning rate of the actor  
LR_CRITIC = 3e-4 # learning rate of the critic  
WEIGHT_DECAY = 0 # L2 weight decay  

In ddpg, the policy is determined by the actor networks and the quality of the policy is dertermined by the critic network. In this perticular implementation the actor NN is made of 3 fully connected layers:
The 1st layer has 24 inputs (the state space of the environment) and 256 outputs
The 2nd layer has 256 inputs and 256 outputs
The final layer has 256 inputs and 2 outputs with tanh activation (One for up-down and one for left-right)

The critic NN is made of 4 fully connected layers: The 1st layer has 24 inputs (the state space of the environment) and 256 outputs
The 2nd layer has 256+2 inputs (previous layer + outputs of actor network) and 256 outputs
The 3rd layer has 256 inputs and 128 outputs
The final layer has 128 inputs and 1 outputs (The value of the policy)

Agent was adapted from https://github.com/udacity/deep-reinforcement-learning/blob/master/ddpg-pendulum/ddpg_agent.py
